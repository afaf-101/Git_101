### Create a k-means estimator and fit the model

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=11)  # find three clusters
kmeans.fit(ws[['Region','Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper',
       'Delicassen']])
```

```python
kmeans2 = KMeans(n_clusters=2, random_state=11)  # find two clusters
kmeans2.fit(ws[['Region','Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper',
       'Delicassen']])
```

### Compare the k-means labels to the whole sale dataset’s target values

```python
print(kmeans.labels_[0:165]) # with three clusters
```

```python
print(kmeans.labels_[165:300]) # with three clusters
```

```python
print(kmeans.labels_[300:440]) # with three clusters
```

```python
print(kmeans2.labels_[0:220]) # with two clusters
```

```python
print(kmeans2.labels_[220:440]) # with two clusters
```

### Dimensionality reduction with Principal Component Analysis (PCA)

```python
ws[['Region','Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper',
       'Delicassen']].shape # shape of features # reduce the 7 features 
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=11)  # reduce to two components
pca.fit(ws[['Region','Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper',
       'Delicassen']])
```

```python
ws_pca = pca.transform(ws[['Region','Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper',
       'Delicassen']])
ws_pca.shape # current shape has been reduced to two so we can visualize
```

### Visualize the reduced data

```python
ws_pca_df = pd.DataFrame(ws_pca, columns=['Component 1', 'Component 2'])
ws_pca_df['Channel'] = ws.Channel # visual channel column values 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
# plot scatter plot of clusters
axes = sns.scatterplot(data=ws_pca_df, hue='Channel', palette='PiYG_r',legend='brief', 
                       x='Component 1', y='Component 2') 

ws_centers = pca.transform(kmeans2.cluster_centers_) 
dots = plt.scatter(ws_centers[:,0], ws_centers[:,1], s=100, c='k') 
```

## Summary/Conclusion

Not much clean up of the dataset was required. We checked to see if there were any null/missing values; there weren't any during the importing, loading and exploration step of the dataset. We used other operations called .value_counts() on the target variable (in this case 'Channel' column), but also did it for the 'Region' column, which we decided to include as part of our feature variables along with the rest of the columns ('Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', and 'Delicassen') to see how many of each type of value was present. Channel column includes information about whether the client of this Portuguese Whole Sale distributor was a hotel/restuaruant/cafe (channel 1) or a retail (channel 2). The Regions column refers to location/region of the client - in this case cities around Portugal like Lisbon (region 1), Oporto (region 2), and other cities (region 3).

The remaining attributes: 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', and 'Delicassen' refer to the annual spending in monetary units (m.u.) on each of these product categories in the country of Portugal. Based on the spending habits of each product category and on the various regions in Portugal, we can classify whether the client of the whole sale distributor is a hotel/restuarant/cafe or a retail. By classifying their clients accordingly, this particular whole sale distributor can expect what sorts of prodcuts to supply to the client based on the type of client and also predict which product categories are popular in what regions of Portugal.

We felt that the dataset matched our labels pretty well. We used three clusters initially for our model and noticed that the centroids seemed pretty close together, so we used two clusters instead. There also seemed to be more overlap between the clusters when we increased it to three. So, yes different numbers of clusters have better isolation of the centroids. We can classify the clusters into two or three depending on how you read the visualization of our reduced data. We can see a clear distinction between the clusters; we can divide it into two-- left, extending up top is one cluster, while the bottom extending to the right is another cluster.
